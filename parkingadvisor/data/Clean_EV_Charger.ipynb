```python
import clean_up as cln
```

```python
# Reads the raw dataset
raw_df = r'.\raw_data\alt_fuel_stations.csv'
col_name = ['Station Name', 'Street Address', 'City', 'ZIP', 'Station Phone',
            'EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count',
            'EV Connector Types', 'Latitude', 'Longitude']
df_ev = cln.read_data(raw_df, col_name)
```

```python
# Filters only Seattle stations
df_ev = df_ev.loc[df_ev['City'] == 'Seattle']
```

```python
# Counts the type of connectors
type_dict = cln.all_type(df_ev)
```

```python
# Adds EV Connector info columns 
cln.ev_connector_tf(df_ev, type_dict)
```

```python
# Adds EV level info columns
cols = ['EV Level1 EVSE Num', 'EV Level2 EVSE Num', 'EV DC Fast Count']
cols_new = ['Level 1', 'Level 2', 'DC Fast']

for level, new_col in zip(cols, cols_new):
    cln.ev_level_tf(df_ev, level, new_col)
```
